```python
query_bucket_mismatch = """
SELECT 
    *,
    CAST(julianday(MaturityDate) - julianday(AsOfDate) AS INTEGER) AS DaysToMaturity
FROM liquidity_data
WHERE 
    (
        LiquidityBucket = '0-7D' AND (julianday(MaturityDate) - julianday(AsOfDate)) > 7
    )
    OR
    (
        LiquidityBucket = '8-30D' AND (julianday(MaturityDate) - julianday(AsOfDate)) NOT BETWEEN 8 AND 30
    )
    OR
    (
        LiquidityBucket = '31-90D' AND (julianday(MaturityDate) - julianday(AsOfDate)) NOT BETWEEN 31 AND 90
    )
    OR
    (
        LiquidityBucket = '91-365D' AND (julianday(MaturityDate) - julianday(AsOfDate)) NOT BETWEEN 91 AND 365
    )
    OR
    (
        LiquidityBucket = '>365D' AND (julianday(MaturityDate) - julianday(AsOfDate)) <= 365
    );
"""

pd.read_sql_query(query_bucket_mismatch, conn)
```

```python
import pandas as pd

# Pull full table from SQLite into a DataFrame
df = pd.read_sql_query("SELECT * FROM liquidity_data;", conn)

df.head()
```

```python
df["AsOfDate"] = pd.to_datetime(df["AsOfDate"])
df["MaturityDate"] = pd.to_datetime(df["MaturityDate"])

df[["AsOfDate", "MaturityDate"]].head()
```

```python
# Calculate days between AsOfDate and MaturityDate
df["DaysToMaturity"] = (df["MaturityDate"] - df["AsOfDate"]).dt.days

# Function to assign the expected bucket based on DaysToMaturity
def assign_bucket(days):
    if days <= 7:
        return "0-7D"
    elif days <= 30:
        return "8-30D"
    elif days <= 90:
        return "31-90D"
    elif days <= 365:
        return "91-365D"
    else:
        return ">365D"

df["ExpectedBucket"] = df["DaysToMaturity"].apply(assign_bucket)

df[["AsOfDate", "MaturityDate", "DaysToMaturity", "LiquidityBucket", "ExpectedBucket"]].head()
```

```python
# 1) Negative Amount
df["DQ_NegativeAmount"] = df["Amount"] < 0

# 2) Missing LiquidityBucket
df["DQ_MissingBucket"] = df["LiquidityBucket"].isna()

# 3) Missing SourceSystem
df["DQ_MissingSourceSystem"] = df["SourceSystem"].isna()

# 4) Bucket mismatch (what we just computed)
df["DQ_BucketMismatch"] = df["LiquidityBucket"] != df["ExpectedBucket"]

# 5) Any DQ issue at all
df["DQ_HasIssue"] = (
    df["DQ_NegativeAmount"] |
    df["DQ_MissingBucket"] |
    df["DQ_MissingSourceSystem"] |
    df["DQ_BucketMismatch"]
)

df.head()
```

```python
# How many rows have each type of issue?
dq_summary = df[[
    "DQ_NegativeAmount",
    "DQ_MissingBucket",
    "DQ_MissingSourceSystem",
    "DQ_BucketMismatch",
    "DQ_HasIssue"
]].sum()

dq_summary
```

```python
# Keep only rows with no DQ issues
df_clean = df[~df["DQ_HasIssue"]].copy()

# Export both full + clean datasets
df.to_csv("liquidity_with_dq_flags.csv", index=False)
df_clean.to_csv("liquidity_clean.csv", index=False)
```

```python
total_rows = len(df)
dq_counts = dq_summary  # from your previous result
dq_pct = (dq_counts / total_rows * 100).round(2)

total_rows, dq_counts, dq_pct
```
